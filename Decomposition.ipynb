## Components

* Systematic : can be decomposed
    * Level: The average value in the series.
    * Trend: The increasing or decreasing value in the series.
    * Seasonality: The repeating short-term cycle in the series.
    
* Non-Systematic : cant be decomposed
    * Noise: The random variation in the series.
    
## Type of Models
* Additive
    * y = level + trend + season + noise
    * Linear model where changes are consistent by same amt over time
    * Trend : straight line
    * Season : same freq and amplitude
* Multiplicative
    * y = level * trend * season * noise
    * Non-Linear model (quad, exponential) where changes dec/inc over time
    * Trend : curved line
    * Season : changes over time - freq and amplitude
   
## Decomposition
* Challenges:
    * Real prob are messy - both additive and multiplicate, inc and dec trends, etc
    * Abstract decompostion gives simple framework to analyze data

```python
# additive decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import numpy as np

a = [i+np.random.randint(10) for i in range(1, 100)]
result = seasonal_decompose(a, model = 'additive', freq = 1) # freq of observations
result.plot()
```

```python
# multiplicative decomposition
a = [i**2 for i in range(1, 100)]
result = seasonal_decompose(a, model = 'additive', freq = 1) # freq of observations
result.plot()
```
